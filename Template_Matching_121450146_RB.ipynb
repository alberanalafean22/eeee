# Templat Matching Korelasi & NCC (cv.TM_CCORR, cv.TM_CCORR_NORMED)

## Template Tidak diputar

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load gambar
img = img
template = tm0

# copy gambar
img_ccorr = img.copy()
img_ccorr_normed = img.copy()

# Ukuran template
w, h = template.shape[::-1]

# Template matching dengan TM_CCORR & TM_CCORR_NORMED
ccorr = cv2.matchTemplate(img, template, cv2.TM_CCORR)
min_val_ccorr, max_val_ccorr, min_loc_ccorr, max_loc_ccorr = cv2.minMaxLoc(ccorr)
ccorr_normed = cv2.matchTemplate(img, template, cv2.TM_CCORR_NORMED)
min_val_ccorr_normed, max_val_ccorr_normed, min_loc_ccorr_normed, max_loc_ccorr_normed = cv2.minMaxLoc(ccorr_normed)

top_left_ccorr = max_loc_ccorr
bottom_right_ccorr = (top_left_ccorr[0] + w, top_left_ccorr[1] + h)
cv2.rectangle(img_ccorr, top_left_ccorr, bottom_right_ccorr, 255, 2)
top_left_ccorr_normed = max_loc_ccorr_normed
bottom_right_ccorr_normed = (top_left_ccorr_normed[0] + w, top_left_ccorr_normed[1] + h)
cv2.rectangle(img_ccorr_normed, top_left_ccorr_normed, bottom_right_ccorr_normed, 255, 2)

# deteksi Templte
plt.figure(figsize=(10, 8))
plt.subplot(221), plt.imshow(img_ccorr, cmap='gray')
plt.title('Detected Template TM_CCORR'), plt.axis('off')
plt.subplot(222), plt.imshow(img_ccorr_normed, cmap='gray')
plt.title('Detected Template TM_CCORR_NORMED'), plt.axis('off')

# matching
plt.subplot(223), plt.imshow(ccorr, cmap='gray')
plt.title('Matching Result TM_CCORR'), plt.axis('off')
plt.subplot(224), plt.imshow(ccorr_normed, cmap='gray')
plt.title('Matching Result TM_CCORR_NORMED'), plt.axis('off')

plt.show()
print(f"Nilai Korelasi Maksimum: {max_val_ccorr_normed}")
```

Hasil Template matching yang dilakukan menggunakan metode korelasi dan NCC, menunjukkan bisa mendeteksi matching dari template terutama pada metode NCC(korelasi dinormalisasi) yang mendeteksi lebih tepat daripada metode korelasi. Hal ini bisa dikarenakan pada NCC data pixel dibaca dinormalisasikan dulu agar seragam. Dari hasil matching dilakukan dengan kedua metode yang menggunakan korelasi, bisa terlihat bahwa template yang memiliki korelasi sangat dekat dengan korelasi maksimum template tidak diputar adalah template 45 derajat. Ketika template matching pada 15 derajat, posisi template sama dengan template tidak diputar, namun ketika pada 30 & 45 derajat posisi template berubah.
